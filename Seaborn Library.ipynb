```python
import seaborn as sns
import numpy as np
import pandas as pd
import pandas.util.testing as tm
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
iris.head()
```

```python
# point plot : which having only single parameters
sns.distplot(iris['petal_length'])
```

```python
#kde= line which shows the max point in the graph
sns.distplot(iris['petal_length'],kde=False)
```

```python
sns.rugplot(iris['petal_length'])
```

```python
# distribution plot : two methods 
# 1. by subplot 2.
sns.residplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
sns.residplot(x='sepal_length',y='sepal_width',data=iris,lowess=True) #lower distribution
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='kde')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='hex',color='#4CB391')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='reg')
```

```python
# pair plot : pair plot shows you to total possible combination of plots 
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')  #  hue parameter determines which column in the data frame should be used for colour encoding.
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tips.head()
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='day')
```

```python
x=sns.pairplot(tips,hue='day')
x.map_diag(sns.distplot)
x.map_upper(sns.kdeplot)
```

```python
 # categorical plot
sns.countplot(x='day',data=tips) #contains only one parametes
```

```python
sns.pointplot(x='total_bill',y='day',data=tips)
```

```python
sns.factorplot(x='total_bill',y='day',data=tips,kind='bar')
```

```python
sns.swarmplot(x='total_bill',y='day',data=tips,hue='sex')
```

```python
sns.swarmplot(x='total_bill',y='day',data=tips,hue='sex',palette='cool')
```

```python
sns.stripplot(x='total_bill',y='day',data=tips,hue='sex',palette='cool')
```

```python
# categorial plot type
# bar plot
sns.barplot(x='day',y='tip',data=tips)
```

```python
# 2. box plot
sns.boxplot(x='day',y='tip',data=tips,hue='sex')
```

```python
# 3. violin plot
sns.violinplot(x='day',y='tip',data=tips,hue='sex')
```

```python
# matrix ploting
# two methods :- 1. Corr 2. pivot
tip2=tips.corr()
```

```python
iris2=iris.corr()
```

```python
sns.heatmap(tip2)
```

```python
sns.heatmap(iris2,annot=True,linewidth=10)
# here values are decreasing
```

```python
# draw matrix graph using pivot method 
tips2=tips.pivot_table(index='day',columns='tip',values='total_bill')
```

```python
tips2
```

```python
sns.heatmap(tips2,lw=1,cmap='cool')
```

```python
sns.heatmap(tips2,cmap='cool')
```

```python
# convert the data and then use it 
# cluster map :
sns.clustermap(iris2,cmap='cool')
```
